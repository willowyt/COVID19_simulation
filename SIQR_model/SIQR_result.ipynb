```python
import numpy as np
import matplotlib.pyplot as plt
from omicron_viral_load import *
```

### R0=9.5

```python
#R=9.5
##lod 3
day=(1,2,3)
L=3
inf=logproportional
N=10000
total3=[]
for D in day:
    St,It,Rt,Qt,SQt,totalIt=SEIRsimulation(N,D,L,inf,asymptomatic=0.90,results_delay=0,R0=9.5,cutoff=6,I_init=10,tmax=90)
    total3=total3+[totalIt]
total3
```

```python
##lod 3---delay
day=(1,2,3)
L=3
inf=logproportional
N=10000
total3_delay=[]
for D in day:
    St,It,Rt,Qt,SQt,totalIt=SEIRsimulation(N,D,L,inf,asymptomatic=0.9,results_delay=1,R0=9.5,cutoff=6,I_init=10,tmax=90)
    total3_delay=total3_delay+[totalIt]
total3_delay
```

```python
##lod 6
L=6
N=10000
D=1
inf=logproportional
N=10000
St,It,Rt,Qt,SQt,total6=SEIRsimulation(N,D,L,inf,asymptomatic=0.90,results_delay=0,R0=9.5,cutoff=6,I_init=10,tmax=90)
total6
```

```python
## no test
N=10000
D=90
L=20
inf=logproportional
St,It,Rt,Qt,SQt,totalIt=SEIRsimulation(N,D,L,inf,asymptomatic=0.90,results_delay=0,R0=9.5,cutoff=6,I_init=10,tmax=90)
totalIt
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 
mpl.rcParams['font.serif'] = ['Microsoft YaHei']
total3=np.array(total3)
total3_delay=np.array(total3_delay)

a=['','','32','31']
bar = np.arange(4)
bar_1=bar[1:]
bar_2 = [i+0.2 for i in bar_1]
bar_3=1/2*(bar_1+bar_2)
bar_3=np.append(0,bar_3)
plt.figure(figsize=(8, 5),dpi=1000)
plt.bar(bar_1, 1-total3/totalIt, width=0.2,color="sienna")
plt.bar(bar_2, 1-total3_delay/totalIt, width=0.2,color="lightcoral")
plt.xticks(bar_3,a,size=10)
plt.bar(0,1-total6/totalIt,width=0.2,color='black')
plt.legend(["","1",""],loc='best')
plt.ylabel('')
plt.ylim(0,1.1)
from matplotlib.ticker import FuncFormatter
def to_percent(temp, position):
  return '%1.0f'%(100*temp) + '%'
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
```
